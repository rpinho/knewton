###Logistical regression of outcome per question per student based on student score on other questions as proxy of student real ability.

```python
import statsmodels.api as sm
import numpy as np
```

```python
questions = pd.DataFrame()
for q_id, score in zip(question_id, loo_score):
    x = score.score
    y = score.correct
    logit = sm.Logit(y, x)
    try:
        result = logit.fit()
    except:
        questions = questions.append({'question_id':q_id, 'logit_coef':np.nan}, ignore_index=True)
    else:
        questions = questions.append({'question_id':q_id, 'logit_coef':result.params[0]}, ignore_index=True)
questions.sort(columns='logit_coef', ascending=False, inplace=True)
```

#Linear Regression

###Create feature matrix with question_id as feature and average score per student as response variable.

```python
# pandas is throwing warning messages
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning, module="pandas", lineno=68)

feature_matrix = df.pivot_table('correct', 'user_id', 'question_id')
feature_matrix#.describe()
```

###Drop students with only 1 valid response

```python
i = feature_matrix.index[(df.groupby('user_id').size() < 2).nonzero()[0]]
feature_matrix = feature_matrix.drop(i)
```

###Drop questions without 0-responses

```python
i = 12665 # all 1s for question_id = 12665
feature_matrix = feature_matrix.drop(i,1)
```

```python
set(feature_matrix)
```

```python
feature_matrix.std()
```

```python
ci = df.groupby('question_id').correct.aggregate(lambda x: np.std(x) / x.count() * 1.96)
```

```python
questions_top50 = ci.index[1:len(ci)/2+2]
#questions_top50 = questions.index[questions.loo_score >= questions.loo_score.median()]
top50 = df.query('question_id in questions_top50')
student_score_top50 = top50.groupby('user_id').correct.mean()
print student_score_top50.describe()
student_score_top50.hist(bins=61)
```
